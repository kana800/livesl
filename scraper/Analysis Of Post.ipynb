```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("sentencizer");
```

#### Helper Functions

```python
from collections import Counter

def mostcommonwords(wrdlist, count):
    """
    summary: return list of most common
    words; The length of the list is determined
    by the count
    args:
        wrdlist -> list
        count -> int
    return 
        list
    """
    word_freq = Counter(wrdlist)
    return word_freq.most_common(count)   

def hashtag_getter(post):
    """summary: returns a list of hashtags
    from the post
    args:
        post -> string
    return:
        list
    """
    return re.findall(r"\B(\#[a-zA-Z]+\b)(?!;)", post)

def removehashtags(post):
    """summary: removes hashtags
    from a given post and return a
    new post
    args:
        post -> string
    return:
        string
    """
    return re.sub(r"\B(\#[a-zA-Z]+\b)(?!;)", "", post)
```

#### Pre-processing content

```python
preprocessedlivepost = []
preprocessednrmpost = []

# saved for hashtag analysis
hashtagslivepost = []
hashtagsnrmpost = []
for post in liveposts:
    hashtagslivepost.append(hashtag_getter(post))
    newpost = removehashtags(post)
    preprocessedlivepost.append(newpost)
    
for post in nrmposts:
    hashtagsnrmpost.append(hashtag_getter(post))
    newpost = removehashtags(post)
    preprocessednrmpost.append(newpost)
```

#### Looking for **most-common** words used in `live-post` events

```python
print("---LIVE EVENT POSTS---")
for post in preprocessedlivepost:
    doc = nlp(post)
    # tokens that arent stop words or punctuations or whitespace
    textwords = [token.text for token in doc
                 if not token.is_stop and not token.is_punct
                 and not token.is_space]
    # tokens that are nouns
    nounwords = [token.text for token in doc
                 if not token.is_stop and not token.is_punct
                 and token.pos_ == "NOUN"]
    # tokens that are verbs
    verbwords = [token.text for token in doc
                 if not token.is_stop and not token.is_punct
                 and not token.is_space and token.pos_ == "VERB"]
    
    for li in [textwords, nounwords, verbwords]:
        mcw = mostcommonwords(li, 5)
        print(mcw)
    print("---")   
```

```python
print("---NORMAL POSTS---")
for post in preprocessednrmpost:
    doc = nlp(post)
    # tokens that arent stop words or punctuations or whitespace
    textwords = [token.text for token in doc
                 if not token.is_stop and not token.is_punct
                 and not token.is_space]
    # tokens that are nouns
    nounwords = [token.text for token in doc
                 if not token.is_stop and not token.is_punct
                 and token.pos_ == "NOUN"]
    # tokens that are verbs
    verbwords = [token.text for token in doc
                 if not token.is_stop and not token.is_punct
                 and not token.is_space and token.pos_ == "VERB"]
    
    for li in [textwords, nounwords, verbwords]:
        mcw = mostcommonwords(li, 5)
        print(mcw)
    print("---")   
```

#### Morphological Analysis
